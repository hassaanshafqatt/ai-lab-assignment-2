```python
from datasets import load_dataset

ds = load_dataset("hassaanshafqat/lang-pop")
```

```python
import numpy as np

df = ds['train'].to_pandas()
df_filtered = df.iloc[::12]

np_array = df_filtered[['Month', 'Python Worldwide(%)', 'JavaScript Worldwide(%)']].to_numpy()

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df_filtered['Month'], df_filtered['Python Worldwide(%)'], label='Python Worldwide(%)')
plt.plot(df_filtered['Month'], df_filtered['JavaScript Worldwide(%)'], label='JavaScript Worldwide(%)')
plt.xlabel('Month')
plt.ylabel('Popularity (%)')
plt.title('Popularity of Python vs JavaScript Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
