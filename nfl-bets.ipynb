```python
import datasets
from datasets import load_dataset


# https://huggingface.co/datasets/tuxmx/nfl_bets_scores
dataset = load_dataset("tuxmx/nfl_bets_scores")

```

```python
import numpy as np

data = dataset['train']
data = data.filter(lambda x: x['score_home'] is not None and x['score_away'] is not None)

scoreshome = data.map(lambda x: {'score_home': x['score_home']})['score_home']

scoresaway = data.map(lambda x: {'score_away': x['score_away']})['score_away']

scores = np.column_stack((scoreshome, scoresaway))

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(scores[:100, 0], label='Home Score')
plt.plot(scores[:100, 1], label='Away Score')
plt.legend()
plt.xlabel('Home Score')
plt.ylabel('Away Score')
plt.title('Home vs Away Scores')
plt.show()
```
