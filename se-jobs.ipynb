```python
from datasets import load_dataset

ds = load_dataset("hassaanshafqat/se-jobs-data")
```

```python
data = ds['train'].to_pandas()

import matplotlib.pyplot as plt

# Filter top 100 companies based on company score
top_100_companies = data.nlargest(100, 'Company Score')

# Plot the graph
plt.figure(figsize=(20, 20))
plt.barh(top_100_companies['Company'], top_100_companies['Company Score'], color='skyblue')
plt.xlabel('Company Score')
plt.ylabel('Company')
plt.title('Top 10 Companies Based on Company Score')
plt.gca().invert_yaxis()
plt.show()
```

```python
average_company_score = data['Company Score'].mean()
print(f"The average company score is: {average_company_score}")
```
