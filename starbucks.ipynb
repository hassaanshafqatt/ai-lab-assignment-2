```python
import datasets
from datasets import load_dataset
import numpy as np


# https://huggingface.co/datasets/hassaanshafqat/starbucks-stock
dataset = load_dataset("hassaanshafqat/starbucks-stock")
```

```python
data = dataset['train']
dates = data['Date']
close = data['Close']

dates_close = np.column_stack((dates, close))




import matplotlib.pyplot as plt

plt.figure(figsize=(60, 6))
plt.plot(dates_close[:50,0], dates_close[:50,1], label='Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Starbucks Stock Close Price Over Time')
plt.legend()
plt.show()
```
